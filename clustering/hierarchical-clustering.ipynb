# Agglomeraitve Clustering

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
data = pd.read_csv('../data/Wholesale customers data.csv')
```

```python
data.head()
```

```python
from sklearn.preprocessing import normalize
data_norm = normalize(data)
data_norm = pd.DataFrame(data=data_norm, columns=data.columns)
data_norm.head()
```

```python
import scipy.cluster.hierarchy as sch

# create the dendrogram 
dend = sch.dendrogram(sch.linkage(data_norm, method='ward'))
plt.figure(figsize=(10,7))
plt.show()
```

```python
# we saw that we have 2 optimal clusters
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(data_norm)
# in the output matrix, 0 and 1 are the labels of the 2 clusters to which the customers belong
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_norm['Milk'], data_norm['Grocery'], c=cluster.labels_)
plt.show()
```
